```python
import pandas as pd
df_students = pd.read_csv('data/grades.csv', delimiter= ',', header='infer')
passed = pd.Series(df_students.Grade>=60)
df_students = pd.concat([df_students, passed.rename('Passed')],axis= 1)
df_students = df_students.dropna(axis=0 )
df_students
```

```python
from matplotlib import pyplot as plt
# fig = plt.figure(figsize=(12.8,7.2))
plt.bar(x = df_students.Name, height= df_students.Grade)
# plt.bar(x = df_students.Name, height= df_students.Grade, color = 'Orange')
# plt.title("Student Grades")
# plt.xlabel("Name")
# plt.ylabel("Grade")
# plt.xticks(rotation=90)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

axes[0].bar(x = df_students.Name, height= df_students.Grade, color = 'Orange')
axes[0].set_title("Student Grades")
axes[0].set_xticklabels(df_students.Name, rotation = 90)

pass_count = df_students.Passed.value_counts()
axes[1].pie(pass_count, labels = pass_count)
axes[1].set_title("Passing Grades")
```
